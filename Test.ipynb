```python
import pandas as pd
import kwantlib as kl
import tskit_learn as tskl

pwrte = pd.read_parquet(r'PWRTE_prices_extract_20240108.parquet')
# offers = pd.read_parquet(r'balancing_capacity_insufficients_offers_extract_20240108.parquet')
# margins = pd.read_parquet(r'balancing_capacity_margins_data_extract_20240108.parquet')
# imbalance = pd.read_parquet(r'balancing_energy_imbalance_data_extract_20240108.parquet')
```

```python
pwrte = pwrte.iloc[:1000, :]
pwrte 
```

```python
prices = pwrte.iloc[:, 1:].sub(pwrte.iloc[:, 0].shift(12), axis=0)
prices
```

```python
signal = kl.proj(kl.cross_moving_average(prices))
signal 
```

```python
kl.Strategy(
    signal = signal.shift(12), 
    returns = prices.diff()
).show()
```
